```python
from odin import math2
from odin import structure
```

```python
cube = zeros((8, 3))
a = 3.07 # inv ang
```

```python
d = array([ a/2, a/2, a/2 ])
cube[0] = d * array([-1,-1,-1])
cube[1] = d * array([-1,-1, 1])
cube[2] = d * array([-1, 1,-1])
cube[3] = d * array([-1, 1, 1])
cube[4] = d * array([ 1,-1,-1])
cube[5] = d * array([ 1,-1, 1])
cube[6] = d * array([ 1,-1,-1])
cube[7] = d * array([ 1, 1, 1])
```

```python
from mpl_toolkits.mplot3d import axes3d, Axes3D

fig = figure()
ax = Axes3D(fig)
ax.scatter(cube[:,0], cube[:,1], cube[:,2], s=100)
show()
```

```python
def on_ewald(xyz, k, delta_q):
    """
    is a mota of diameter `delta_q` at `xyz` on an ewald sphere (of radius `k`)
    """
    
    assert xyz.shape == (3,)
    
    ewald_center = array([ 0.0, 0.0, -k ])
    dist = abs(np.linalg.norm( xyz - ewald_center ) - k)
    
    if dist < delta_q:
        is_on_ewald = 1
    else:
        is_on_ewald = 0
    
    return is_on_ewald


def dist_to_ewald(xyz, k, delta_q):
    assert xyz.shape == (3,)
    ewald_center = array([ 0.0, 0.0, -k ])
    dist = abs(np.linalg.norm( xyz - ewald_center ) - k)
    return dist
```

```python
n_samples = 100000
k = 2. * pi / 0.7293
delta_q = 0.0710 / 2.
n_intersect = zeros(n_samples, dtype=int)

ds = zeros(n_samples)
for i in range(n_samples):
    rotated_cube = structure.rand_rotate_molecule2(cube)
    n_intersect[i] = sum( [on_ewald(rotated_cube[j], k, delta_q) for j in range(8)] )
    ds[i] = dist_to_ewald(rotated_cube[0], k, delta_q)
    
#hist(ds, 100)
```

```python
fig = figure()
ax = Axes3D(fig)
ax.scatter(rotated_cube[:,0], rotated_cube[:,1], rotated_cube[:,2], s=1000)
show()

hist(n_intersect, bins=[0,1,2,3,4], normed=True)
xlabel('Number of Bragg Intersections')
ylabel('Frequency')
```

```python
intersection_count = bincount(n_intersect)
print "Fraction of single reflections: %.1f%%" % (float(intersection_count[1]) / float(n_samples) * 100.0)
print "Fraction of double reflections: %.1f%%" % (float(intersection_count[2]) / float(n_samples) * 100.0)
print "Fraction of triple reflections: %.1f%%" % (float(intersection_count[3]) / float(n_samples) * 100.0)
print "Number of particles with one or more intersection: %.1f%%" % (float(sum(intersection_count[1:])) / float(n_samples) * 100.0)
```
